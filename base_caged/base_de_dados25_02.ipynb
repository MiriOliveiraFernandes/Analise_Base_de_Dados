```python
from google.colab import files
import pandas as pd

uploaded = files.upload()
df = pd.read_excel("base_dados_caged_evSalario.xlsx")
df.head()
```

```python
print(df.columns)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

**Varificando o maior valor de admissão**

```python
indice_maior_valor_adm= df["Salário Médio Real de Admissão"].idxmax()
maior_valor_admissao = round(df.loc[indice_maior_valor_adm,"Salário Médio Real de Admissão"],2)
mes_correspondente = df.loc[indice_maior_valor_adm,"Mês"]
print(f"Mês Correspondete: {mes_correspondente}\nMaior valor de Salário Médio Real de Admissão R$:{maior_valor_admissao}")
```

**Verificando os Top 10 Salário Médio de ADMISSÃO**





```python
top_10_maiores = df.nlargest(10,"Salário Médio Real de Admissão")
top_10_maiores["Salário Médio Real de Admissão"] = top_10_maiores["Salário Médio Real de Admissão"].round(2)
print(top_10_maiores[["Salário Médio Real de Admissão","Mês"]])
```

**Criando um grafico com os TOP-10 salário de ADMISSÃO**





```python
plt.figure(figsize=(8,5))
#cria um grafico de linhas
sns.lineplot(x=top_10_maiores["Mês"], y=top_10_maiores["Salário Médio Real de Admissão"], marker = "o")
plt.title("Top 10 Salários Média Real de Admissão")
plt.xlabel("Mês")
plt.ylabel("Salário")
plt.xticks(rotation=45)
plt.show()
```

**Verifica o maior valor do Salário de DESLIGAMENTO**

```python
indice_maior_valor_desligamento = df["Salário Médio Real de Desligamento"].idxmax()
maior_valor_delisgamento = round(df.loc[indice_maior_valor_desligamento,"Salário Médio Real de Desligamento"],2)
mes_correspondente_desligamento = df.loc[indice_maior_valor_desligamento,"Mês"]
print(f"Mês Correspondente: {mes_correspondente_desligamento}\nMaior valor de Salário Médio Real de Desligamento R$:{maior_valor_delisgamento}")
```

**Verificando os TOP-10 salário de DESLIGAMENTO**

```python
top_10_maiores_desligamento = df.nlargest(10,"Salário Médio Real de Desligamento")
top_10_maiores_desligamento["Salário Médio Real de Desligamento"] = top_10_maiores_desligamento["Salário Médio Real de Desligamento"].round(2)
print(top_10_maiores_desligamento[["Salário Médio Real de Desligamento","Mês"]])
```

**Criando um grafico com os TOP-10 salário de DESLIGAMENTO**

```python
plt.figure(figsize=(10,6))
sns.lineplot(x=top_10_maiores_desligamento["Mês"], y=top_10_maiores_desligamento["Salário Médio Real de Desligamento"], marker = "o",color='y')
plt.title("Top 10 Salário Médio do Desligamento")
plt.xlabel("Mês")
plt.ylabel("Salário")
plt.xticks(rotation = 45)
plt.show()
```

**Cálculo da diferença entre os salários**

```python
maior_salario_admissao = round(df["Salário Médio Real de Admissão"].max(),2)
menor_salario_admissao = round(df["Salário Médio Real de Admissão"].min(),2)
diferenca_salario_adm = round(maior_salario_admissao - menor_salario_admissao,2)

print(maior_salario_admissao)
print(menor_salario_admissao)
print(f"Diferença entre os salrios de admissão {diferenca_salario_adm}")
```

```python
maior_salario_desligamento = round(df["Salário Médio Real de Desligamento"].max(),2)
menor_salario_desligamento = round(df["Salário Médio Real de Desligamento"].min(),2)
diferenca_salario_desligamento = round(maior_salario_desligamento - menor_salario_desligamento,2)

print(f"Maior Salário de Desligamento R$:{maior_salario_desligamento}")
print(f"Menor Salário de Desligamento R$:{menor_salario_desligamento}")
print(f"Diferença entre os salários de desligamento {diferenca_salario_desligamento}")
```
